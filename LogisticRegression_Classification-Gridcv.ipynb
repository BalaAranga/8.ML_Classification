```python
import pandas as pd
```

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
dataset
```

```python
dataset= pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset = dataset.drop("User ID",axis=1)
dataset
```

```python
dataset["Purchased"].value_counts()
```

```python
independent = dataset[['Age', 'EstimatedSalary','Gender_Male']]
independent.shape                   
```

```python
dependent = dataset[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=1/3,random_state=0) 
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
param_grid = {'penalty' :['l2', 'elasticnet', 'none'],
              'C' :[0.1,0.03,0.05,1,1.5,2],
              'fit_intercept':['True', 'False'],
              'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
              'multi_class':['auto', 'ovr', 'multinomial']}
grid = GridSearchCV(LogisticRegression(),param_grid, refit=True, verbose=3, n_jobs=1)
grid.fit(x_train,y_train) 
```

```python
grid_prediction = grid.predict(x_test) # AI Brain giving output(predicting)
grid_prediction
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,grid_prediction)
print("The confusion_matrix Score Value for Best Parameter {}:".format(grid.best_params_),cm)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,grid_prediction)
```

```python
# LogisticRegression
print(clf_report)
```
